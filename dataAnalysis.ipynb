```python
def get_data_structure(start_day,end_day,limit):  # 构建列表生成器
    pipeline=[
    {'$match':{'$and':[{'pub_date':{'$gte':start_day,'$lte':end_day}},{'area':{'$all':['朝阳']}}]}},
    # 筛选出2015.12.25——2015.12.27时段，朝阳区发出的帖子数量
    {'$group':{'_id':{'$slice':['$cates',2,1]},'count':{'$sum':1}}},
    # _id表示分组的因子，根据商品种类进行分组，用count来统计数目，$sum表示每找到一个加1
    {'$sort':{'count':-1}},
    # 根据count数倒序输出，1为顺序，－1为倒序
    {'$limit':limit}
    # 表示只输出三位元素
    ]
    for i in item_info.aggregate(pipeline):
        data={  # 构建画图所需要的数据结构
            'name':i['_id'][0],
            'type':'column',
            'data':[i['count']]
        }
        yield data  # yield表示每循环一次就中断一次，输出一个值
        
for i in get_data_structure('2015.12.25','2015.12.27',3):
    print(i)
```

```python
series=[i for i in get_data_structure('2015.12.25','2015.12.27',5)] # 选取排行前五的类目
option={
    'chart':{'zoomType':'xy'}, 
    'title'   : {'text': '发帖数量最大的类目'}, # 绘图的标题
    'subtitle': {'text': '数据图表'}, # 绘图的副标题
    'yAxis'   : {'title': {'text': '数量'}} # y轴的名称
    }
charts.plot(series,options=option,show='inline')
```

```python
for i in item_info.find():
    if not(isinstance(i['price'],int)) and list(i['price'].split(' '))[0] != '面议' :
        if i['price'] == '${info.paramsMap.minprice}':
            price = '面议'
        else:
            price = list(i['price'].split(' '))[0] # 将价格转化为整数
            if ',' in price: # 合并442，22 逗号
                price = int(''.join(price.strip().split(',')))
            elif '元' in price: # 考虑到160元，未以空格分开的情况
                price = int(list(price.split('元'))[0])
            else:
                price = int(price)
    else:   
        price = i['price']
    item_info.update({'_id':i['_id']},{'$set':{'price':price}}) # update database
```

```python
pipeline_price=[
    {'$match':{'$and':[
                {'pub_date':{'$gte':'2015.12.25','$lte':'2015.12.27'}}, # 发帖时间段
                {'cates':{'$all':['北京二手手机']}}, # 发帖种类
                {'look':{'$nin':['-']}}, # 商品新旧程度
                {'price':{'$nin':['面议']}} # 去除未知价格
                 ]}},
    {'$group':{'_id':'$look','average_price':{'$avg':'$price'}}}, # 根据新旧程度分组，计算组间平均价格
    {'$sort':{'average_price':-1}} # 根据平均价格到序输出    
]
for i in item_info.aggregate(pipeline_price):
    print(i)
    
```

```python
def get_lookData(start_day,end_day,category):
    pipeline_price=[
    {'$match':{'$and':[
                {'pub_date':{'$gte':start_day,'$lte':end_day}}, # 发帖时间段
                {'cates':{'$all':[category]}}, # 发帖种类
                {'look':{'$nin':['-']}} # 商品新旧程度
                 ]}},
    {'$group':{'_id':'$look','average_price':{'$avg':'$price'}}}, # 根据新旧程度分组，计算组间平均价格
    {'$sort':{'average_price':1}} # 根据平均价格顺序输出    
    ]
    for i in item_info.aggregate(pipeline_price):
        yield i['average_price']       
series = [
    {'name':'北京二手手机', 
    'type':'line',
    'data':[i for i in get_lookData('2015.12.25','2015.12.27','北京二手手机')]},
    {'name':'北京二手笔记本',
    'type':'line',
    'data':[i for i in get_lookData('2015.12.25','2015.12.27','北京二手笔记本')]}]
options={
    'title': {'text': '新旧-价格'},
    'xAxis'   : {'categories': ['报废机/尸体','7成新及以下','8成新','9成新','95成新','99成新', '全新']},
    'yAxis'   : {'title': {'text': '价格'}},
}
charts.plot(series,options=options,show = 'inline')
```
